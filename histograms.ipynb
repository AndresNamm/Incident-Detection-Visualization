```python
import numpy as np
import matplotlib.pyplot as plt 

t = []
k=[]
c=[]
l=[]

for i in range(10):
    t.append(i)
    k.append(i)
 
for i in range(10):
    c.append(i*2)
    l.append(i*10)
      

plt.plot(t,k)
plt.plot(c,l)
points= [[1,20], [12.5,50]]
x=[]
y=[]
for i in points:
    x.append(i[0])
    y.append(i[1])
    
plt.plot(x,y,'ro')
x_polated= x[0]+ 0.75*(x[1]-x[0]) 
y_polated= y[0]+ 0.75*(y[1]-y[0])        

plt.plot(x_polated,y_polated,'ro')         
plt.show()
```

# HISTOGRAM OF TRAJECTORY  LENGTHS 


```python
import matplotlib.pyplot as plt
import numpy as np

f = open("trajHistogram.txt")
cnts =[]

for i in f:
    cnts.append(int(i))


start = 1
end = 4
x = [i for i in range(start,end,1)]
y = np.array(cnts)
plt.hist(y, range=[start,end],bins=end-start )
plt.ylabel('Nr of Trajectories having x record count');
plt.xticks(x)
plt.show()
f.close()

```

```python
import matplotlib.pyplot as plt
import numpy as np

f_new = open("difHistNew.txt")
f_old = open("difHistOld.txt")
cnts_new =[]
cnts_old =[]
for i in f_new:
    cnts_new.append(int(i))
for i in f_old:
    cnts_old.append(int(i))



start = 2
end = 4
x = [i for i in range(start,end,1)]
y = np.array(cnts_new)
y_old=np.array(cnts_old)
#plt.figure()
#plt.subplot(211)
plt.ylabel('Nr of Trajectories having x record count');
plt.hist(y, range=[start,end],bins=end-start,normed=False ,label = "new")
plt.hist(y_old,  range=[start,end],bins=end-start,normed=False, label ="old")
plt.xticks(x)
plt.legend(loc='upper right')

plt.show()


start=5
end=20
x = [i for i in range(start,end,1)]
y = np.array(cnts_new)
y_old=np.array(cnts_old)
#plt.figure()
#plt.subplot(212)
plt.ylabel('Nr of Trajectories having x record count');
plt.hist(y, range=[start,end],bins=end-start )
plt.hist(y_old,  range=[start,end],bins=end-start,normed=False, label ="old")
plt.legend(loc='upper right')
plt.xticks(x)



#plt.hist(y,bins=bin)
plt.show()
f_new.close()
```

```python


import matplotlib.pyplot as plt
import numpy as np

f_old = open("old.csv")
f_new_gps = open("gps.csv")
f_new_sp = open("gps_sp.csv")
cnts_new_sp =[]
cnts_new_gps =[]
cnts_old =[]
for i in f_new_gps:
    cnts_new_gps.append(float(i))
for i in f_old:
    cnts_old.append(float(i))
for i in f_new_sp:
    cnts_new_sp.append(float(i))

start = 0
end =1
y_gps = np.array(cnts_new_gps)
y_old=np.array(cnts_old)
y_sp  = np.array(cnts_new_sp)
#plt.figure()
#plt.subplot(211)
plt.ylabel('Histogram for AUC values');
bins = 45
plt.hist(y_sp,  range=[start,end],bins=bins,alpha=0.5,normed=False,label="gps+sp",)
plt.hist(y_gps, range=[start, end],bins=bins,alpha=0.5, normed=False, label="gps")
plt.hist(y_old,  range=[start,end],bins=bins,alpha=1,normed=False, label="old proj")

plt.legend(loc='upper right')

plt.show()
f_old.close()
f_new_gps.close()
f_new_sp.close()






```
